```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
img=cv2.imread("kids.jpg")
newImg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.imshow(newImg)
plt.axis("off")
plt.show()
```

```python
model=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=model.detectMultiScale(newImg,1.3,1)
```

```python
print(faces)
```

```python
for face in faces:
  x,y,z,w=face
  cv2.rectangle(newImg,(x,y),(x+z,y+w),(0,255,0),2)
plt.imshow(newImg)
plt.show()
```
