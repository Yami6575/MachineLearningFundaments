```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("train.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
mu=x.mean(axis=0)
std=x.std(axis=0)
X=(x-mu)/std
X
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features=SelectKBest()
```

```python
fit=best_features.fit(X,y)
fit.scores_
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcolumn=pd.DataFrame(X.columns)
```

```python
featurescores=pd.concat([dfcolumn,dfscores],axis=1)
featurescores.columns=['Features','Scores']
featurescores
```

```python
featurescores=featurescores.sort_values(by='Scores',ascending=False)
featurescores
```

```python
top_10=list(featurescores.iloc[:10]['Features'].values)
top_10
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
scores = cross_val_score(model, X, y, cv = 10)
scores.mean()
```

```python
X[top_10]
```

```python
scores = cross_val_score(model, X[top_10], y, cv = 10)
scores.mean()
```
