```python
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
import numpy as np
```

```python
image=cv2.imread("download.jpeg")
print(image.shape)
```

```python
plt.imshow(image)
plt.show()
```

```python
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
img=cv2.resize(img,(382,306))
plt.show()
```

```python
img[0]
```

```python
img=img.reshape(-1,3)
```

```python
img
```

```python
k=10
model=KMeans(n_clusters=k)
model.fit(img)
```

```python
centroids=model.cluster_centers_
```

```python
print(centroids)
```

```python
colors=np.array(centroids,dtype='uint8')
```

```python
colors
```

```python
i=1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis("off")
    i=i+1
    mat=np.zeros((100,100,3),dtype='uint8')
    mat[:,:,:]=color
    plt.imshow(mat)
    plt.show()
```
